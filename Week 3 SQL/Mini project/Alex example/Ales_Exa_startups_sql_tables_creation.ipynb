```python
import pandas as pd
```

```python
su_df = pd.read_csv('/Users/miriamg/Documents/GitHub_Ironhack/Week 3 SQL/Mini project/startups.csv')
```

```python
su_df.dropna(inplace=True)
su_df.reset_index(inplace=True)
```

```python
su_df
```

```python
countries = sorted(list(su_df['country'].unique()))
countries
```

```python
country_id = list(range(1,len(countries)+1))
country_id
```

## Create country table

```python
country_mapping = dict(zip(countries,country_id))
country_mapping
```

```python
su_df['country_id'] = su_df['country'].map(country_mapping)
```

```python
su_df['country_id']
```

```python
country_dict = {'country_id':country_id, 'country_name':countries}
country_df = pd.DataFrame(country_dict)
country_df
```

```python
country_df.to_csv("countries.csv")
```

## Create the city table

```python
city_df = su_df[['city','country_id']]
```

```python
city_df.drop_duplicates(inplace=True)
```

```python
city_df.reset_index(inplace=True,drop=True)
```

```python
city_df = city_df.sort_values(by='city')
```

```python
city_df.reset_index(inplace=True,drop=True)
```

```python
city_df['city_id'] = list(range(1,len(city_df)+1))
```

```python
city_df
```

```python
cities = sorted(list(su_df['city'].unique()))
city_id = list(range(1,len(cities)+1))
```

```python
city_dict = {'city_id':city_id, 'city_name':cities}
city_df = pd.DataFrame(city_dict)
city_df
```

```python
su_df['city_id'] = su_df['city'].map(dict(zip(cities,city_id)))

su_df
```

## Create the industries table

```python
industries = sorted(list(su_df['industry'].unique()))
```

```python
industry_df = pd.DataFrame({'industry_id': list(range(1,len(industries)+1)),'industry_name':industries})
```

```python
industry_df
```

```python
su_df['industry_id'] = su_df['industry'].map(dict(zip(industries,list(range(1,len(industries)+1)))))
```

## Finally, create the startups table

```python
startups_df = su_df[['company','valuation_($b)','date_joined','city_id','industry_id']]
startups_df = startups_df.drop_duplicates()
```

```python
startups_df
```

```python
startups_df.insert(0,'startup_id',list(range(1,len(startups_df)+1)))
```

```python
startups_df
```

## Create csv's

```python
country_df.to_csv('country_table.csv')
city_df.to_csv('city_table.csv')
startups_df.to_csv('startups_table.csv')
industry_df.to_csv('industry_table.csv')
```
