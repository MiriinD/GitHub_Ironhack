In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here

import pandas as pd

url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df1 = pd.read_csv(url1)
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
df2 = pd.read_csv(url2)
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
df3 = pd.read_csv(url3)
```

```python
df1.columns
```

```python
df2.columns
```

```python
df3.columns
```

```python
df3= df3.rename(columns= {"Gender": "GENDER"})
df3
df3= df3.rename(columns= {"State": "ST"})
df3
```

```python
comb1_df= pd.concat([df1, df2, df3], axis=0, ignore_index=True)

comb1_df
```

```python
lc_separate_columns_df = comb1_df.rename(columns=lambda x: x.lower().replace(" ", "_"))


clean_columns_df= lc_separate_columns_df.rename(columns={"st": "state"})

clean_columns_df["gender"] = clean_columns_df["gender"].replace(["Femal", "female"], "F").replace(["Male", "male"], "M")
clean_columns_df["state"] = clean_columns_df["state"].replace("AZ", "Arizona").replace("Cali", "California").replace("WA", "Washington")
clean_columns_df["education"] = clean_columns_df["education"].replace("Bachelors", "Bachelor")
clean_columns_df["customer_lifetime_value"] = clean_columns_df["customer_lifetime_value"].str.replace('%', '')
clean_columns_df["vehicle_class"] = clean_columns_df["vehicle_class"].replace(["Sports Car", "Luxury SUV", "Luxury Car"], "Luxury")
clean_columns_df["customer_lifetime_value"] = clean_columns_df["customer_lifetime_value"].astype(float)
clean_columns_df["number_of_open_complaints"] = clean_columns_df["number_of_open_complaints"].apply(lambda x: int(x.split("/")[1]) if type (x)== str else x)
clean_columns_df
```

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here
url4 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"
df4 = pd.read_csv(url4)
df4.info()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_marketing = df4.pivot_table(index='sales_channel', values='monthly_premium_auto', aggfunc='sum').round(2)
pivot_marketing #Agent and branch represent the highest revenue, this could be bcs they have more clients (Str to number to add others)
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_revenue_gender_education = df4.pivot_table(index='education', columns='gender', values='customer_lifetime_value', aggfunc=('mean', 'median'))
pivot_revenue_gender_education #The higher the education of customers, the lower the revenues?
```
